# AI Agents and LLM Workflow Design Patterns

This notebook introduces the concept of **AI Agents** and explores the five major **workflow design patterns** for building systems with Large Language Models (LLMs).  
The content is structured in a **Q&A format** for clarity, with detailed explanations, key characteristics, and practical examples.  

---

## Part 1: AI Agents

### Question  
**What are AI Agents?**

---

### Answer  

**AI Agents** are intelligent software programs powered by **Large Language Models (LLMs)** that can **reason, plan, and take actions**.  
Unlike simple chat models that only generate text, AI Agents are capable of:  

- **Planning tasks** ‚Üí Identifying the steps required to achieve a goal.  
- **Calling tools and APIs** ‚Üí Using external systems such as search engines, databases, or calculators.  
- **Managing actions** ‚Üí Executing workflows, monitoring progress, and adapting based on feedback.  

In short, AI Agents combine **language understanding** with **decision-making and action execution**, enabling them to solve real-world problems autonomously.  

---

### Key Characteristics of AI Agents  

- **Autonomous** ‚Üí Can act without step-by-step human supervision.  
- **Goal-Oriented** ‚Üí Focused on reaching a defined objective.  
- **Tool-Using** ‚Üí Can interact with APIs, software, and data sources.  
- **Adaptive** ‚Üí Adjust steps based on context or feedback.  
- **Workflow-Aware** ‚Üí Break down and manage multi-step processes.  

---

### Example  

**User Request:**  
‚ÄúBook me a flight to Delhi next week for under $200.‚Äù  

**How an AI Agent responds:**  
1. **Plan** ‚Üí Identify required steps: search flights ‚Üí filter ‚Üí compare ‚Üí book.  
2. **Call Tools** ‚Üí Query a flight search API.  
3. **Evaluate** ‚Üí Check results against price and schedule constraints.  
4. **Act** ‚Üí Reserve the best option and confirm the booking.  

This demonstrates that an AI Agent is not limited to text generation.  
It operates as a decision-maker and executor of actions in a workflow.

## Part 2: LLM Workflow Design Patterns  

LLMs are often combined into **workflows** to improve efficiency, accuracy, and control.  
Anthropic identifies **five major workflow design patterns** commonly used when building generative AI systems.  

---

### 1. Prompt Chaining  

**Definition:**  
A sequence of LLM calls where the output of one step becomes the input of the next.  

**Example:**  
- LLM 1 ‚Üí Selects a business sector.  
- LLM 2 ‚Üí Identifies a pain point.  
- LLM 3 ‚Üí Suggests a solution.  

**Key Characteristics:**  
- Breaks a complex task into smaller, well-defined subtasks.  
- Provides structure and keeps the process on guardrails.  
- Flexible: can use any number of chained LLMs.  

---

### 2. Routing  

**Definition:**  
An LLM acts as a ‚Äúrouter,‚Äù deciding which specialized model should process the input.  

**Example:**  
- Router receives a query.  
- Sends math problems ‚Üí Math-specialist LLM.  
- Sends legal questions ‚Üí Legal-specialist LLM.  

**Key Characteristics:**  
- Efficient use of specialized expertise.  
- Input classification handled by the router LLM.  
- Provides autonomy within controlled boundaries.  

---

### 3. Parallelization  

**Definition:**  
A task is divided into multiple subtasks and executed simultaneously.  
Here, **code (not an LLM)** controls task splitting and result aggregation.  

**Example:**  
- Summarizing a 5-chapter book.  
- Each chapter sent to a different LLM in parallel.  
- Results combined into one summary.  

**Key Characteristics:**  
- Improves efficiency by running tasks concurrently.  
- Can repeat the same task for robustness (e.g., majority voting).  
- Aggregation handled programmatically.  

---

### 4. Orchestrator-Worker  

**Definition:**  
An **LLM orchestrator** manages the workflow by assigning subtasks to **worker LLMs** and synthesizing their outputs.  

**Example:**  
- Orchestrator LLM splits essay into: introduction, body, conclusion.  
- Worker LLMs generate each section.  
- Orchestrator merges results into a final essay.  

**Key Characteristics:**  
- Similar to parallelization but controlled by an LLM instead of code.  
- More dynamic and adaptive.  
- Closer to an agent pattern, but still constrained.  

---

### 5. Evaluator-Optimizer  

**Definition:**  
A two-LLM pattern where one generates outputs and another evaluates them.  

**Example:**  
- Generator LLM writes an essay.  
- Evaluator LLM reviews it for grammar, accuracy, and logic.  
- If accepted ‚Üí Output finalized.  
- If rejected ‚Üí Feedback returned ‚Üí Generator improves the result.  

**Key Characteristics:**  
- Feedback loop improves quality and reliability.  
- Increases accuracy and predictability.  
- Widely used in production systems.  

---

## Part 3: Quick Recap  

- **Prompt Chaining** ‚Üí Step-by-step workflow.  
- **Routing** ‚Üí Choose the right expert model.  
- **Parallelization** ‚Üí Tasks executed at the same time (managed by code).  
- **Orchestrator-Worker** ‚Üí LLM manages workers dynamically.  
- **Evaluator-Optimizer** ‚Üí Generator + Evaluator loop for refinement.  

---

## Final Takeaway  

- **AI Agents** extend LLMs with reasoning, planning, and action-taking abilities.  
- **Workflow patterns** provide structured methods to coordinate LLMs effectively.  
- Together, they enable the creation of powerful AI systems that are accurate, efficient, and suitable for real-world applications.  


# üîÑ 5 Workflow Design Patterns with LLMs

Large Language Model (LLM) workflows can be structured into specific design patterns that help manage complexity, efficiency, and reliability.  
Here are **five key workflow patterns** with detailed explanations.

---

## 1. Prompt Chaining  
**Definition:**  
Breaking a task into **fixed sub-tasks**, where each LLM output is passed into the next stage.  

- **Use Case:**  
  - Step-by-step reasoning  
  - Report generation  
  - Multi-stage data transformation  

- **Advantage:**  
  - Predictable execution  
  - Easier debugging and monitoring  

- **Limitation:**  
  - Rigid structure (not adaptive)  
  - Errors in early steps propagate  

**Flow:**  


```python
from graphviz import Digraph

# 1. Prompt Chaining
dot1 = Digraph(comment="Prompt Chaining")
dot1.attr(rankdir="LR")
dot1.node("IN", shape="circle", style="filled", color="orange")
dot1.node("LLM1", shape="box", style="filled", color="gold")
dot1.node("Gate", shape="box", style="filled", color="lightblue")
dot1.node("LLM2", shape="box", style="filled", color="gold")
dot1.node("LLM3", shape="box", style="filled", color="gold")
dot1.node("OUT", shape="circle", style="filled", color="orange")

dot1.edges([("IN", "LLM1"), ("LLM1", "Gate"), ("Gate", "LLM2"), ("LLM2", "LLM3"), ("LLM3", "OUT")])
dot1
```


---

## 2. Routing  
**Definition:**  
Inputs are **directed into specialized sub-tasks** using a router model that decides the best path.  

- **Use Case:**  
  - Customer support (billing vs. technical)  
  - Domain-specific assistants  

- **Advantage:**  
  - Separation of concerns  
  - Efficiency through specialization  

- **Limitation:**  
  - Router must be accurate  
  - Wrong routing degrades performance  

**Flow:**  


```python
# 2. Routing
dot2 = Digraph(comment="Routing")
dot2.attr(rankdir="LR")
dot2.node("IN", shape="circle", style="filled", color="orange")
dot2.node("Router", "LLM Router", shape="box", style="filled", color="lightblue")
dot2.node("LLM1", shape="box", style="filled", color="gold")
dot2.node("LLM2", shape="box", style="filled", color="gold")
dot2.node("LLM3", shape="box", style="filled", color="gold")
dot2.node("OUT", shape="circle", style="filled", color="orange")

dot2.edge("IN", "Router")
dot2.edge("Router", "LLM1")
dot2.edge("Router", "LLM2")
dot2.edge("Router", "LLM3")
dot2.edges([("LLM1", "OUT"), ("LLM2", "OUT"), ("LLM3", "OUT")])
dot2
```


---

## 3. Parallelization  
**Definition:**  
Task is **split into multiple subtasks**, executed concurrently, and results are aggregated.  

- **Use Case:**  
  - Summarizing multiple documents  
  - Parallel Q&A across data sources  

- **Advantage:**  
  - Faster execution  
  - Scalable with resources  

- **Limitation:**  
  - Aggregation logic is non-trivial  
  - Consistency across outputs required  

**Flow:**  


```python
# 3. Parallelization
dot3 = Digraph(comment="Parallelization")
dot3.attr(rankdir="LR")
dot3.node("IN", shape="circle", style="filled", color="orange")
dot3.node("Coord", "Coordinator", shape="box", style="filled", color="lightblue")
dot3.node("LLM1", shape="box", style="filled", color="gold")
dot3.node("LLM2", shape="box", style="filled", color="gold")
dot3.node("LLM3", shape="box", style="filled", color="gold")
dot3.node("Agg", "Aggregator", shape="box", style="filled", color="lightblue")
dot3.node("OUT", shape="circle", style="filled", color="orange")

dot3.edge("IN", "Coord")
dot3.edge("Coord", "LLM1")
dot3.edge("Coord", "LLM2")
dot3.edge("Coord", "LLM3")
dot3.edge("LLM1", "Agg")
dot3.edge("LLM2", "Agg")
dot3.edge("LLM3", "Agg")
dot3.edge("Agg", "OUT")
dot3
```


---

## 4. Orchestrator-Worker  
**Definition:**  
An **Orchestrator dynamically breaks down** a complex task, assigns it to workers, and then synthesizes results.  

- **Use Case:**  
  - Research agents  
  - Planning and execution frameworks  

- **Advantage:**  
  - Handles complex, dynamic workflows  
  - Adaptive to task requirements  

- **Limitation:**  
  - Computationally expensive  
  - Requires reliable orchestration logic  

**Flow:**  


```python
# 4. Orchestrator-Worker
dot4 = Digraph(comment="Orchestrator-Worker")
dot4.attr(rankdir="LR")
dot4.node("IN", shape="circle", style="filled", color="orange")
dot4.node("Orch", "Orchestrator", shape="box", style="filled", color="lightblue")
dot4.node("LLM1", shape="box", style="filled", color="gold")
dot4.node("LLM2", shape="box", style="filled", color="gold")
dot4.node("LLM3", shape="box", style="filled", color="gold")
dot4.node("Synth", "Synthesizer", shape="box", style="filled", color="lightblue")
dot4.node("OUT", shape="circle", style="filled", color="orange")

dot4.edge("IN", "Orch")
dot4.edge("Orch", "LLM1")
dot4.edge("Orch", "LLM2")
dot4.edge("Orch", "LLM3")
dot4.edge("LLM1", "Synth")
dot4.edge("LLM2", "Synth")
dot4.edge("LLM3", "Synth")
dot4.edge("Synth", "OUT")
dot4
```


---

## 5. Evaluator-Optimizer  
**Definition:**  
One LLM **generates an answer**, and another LLM **evaluates/validates** it. If rejected, the generator retries with feedback.  

- **Use Case:**  
  - Code generation and debugging  
  - High-stakes domains (law, medicine, finance)  

- **Advantage:**  
  - Produces higher quality, validated results  
  - Improves safety and reliability  

- **Limitation:**  
  - Slow due to multiple passes  
  - Costly in compute resources  

**Flow:**  


```python
# 5. Evaluator-Optimizer
dot5 = Digraph(comment="Evaluator-Optimizer")
dot5.attr(rankdir="LR")
dot5.node("IN", shape="circle", style="filled", color="orange")
dot5.node("Gen", "LLM Generator", shape="box", style="filled", color="gold")
dot5.node("Eval", "LLM Evaluator", shape="box", style="filled", color="lightblue")
dot5.node("OUT", shape="circle", style="filled", color="orange")

dot5.edge("IN", "Gen")
dot5.edge("Gen", "Eval")
dot5.edge("Eval", "OUT", label="Accepted", color="green")
dot5.edge("Eval", "Gen", label="Rejected w/ feedback", color="red")
dot5
```
